```python
"""Plate detection module using YOLOv8."""

from pathlib import Path

import wandb
from ultralytics import YOLO
```

```python
def main():
    """Train YOLOv8 model for Malaysian license plate detection."""
    dataset_path = Path("../plate-dataset/data.yaml")

    config = {
        "model": "yolov8s.pt",
        "epochs": 100,
        "img_size": 640,
        "batch_size": 8,
        "device": "mps",
        "patience": 20,
    }

    wandb.init(
        project="plate-detector-model",
        name="v5",
        config=config,
    )

    model = YOLO(config["model"])

    model.train(
        data=str(dataset_path),
        epochs=config["epochs"],
        imgsz=config["img_size"],
        batch=config["batch_size"],
        device=config["device"],
        patience=config["patience"],
        project="plate_detection",
        name="malaysia_plate_v5",
        exist_ok=True,
    )

    best_model_path = Path(
        "runs/plate_detection/malaysia_plate_v5/weights/best.pt"
    )

    if best_model_path.exists():
        wandb.save(str(best_model_path))

    wandb.finish()


if __name__ == "__main__":
    main()
```

```python
model = YOLO("runs/detect/plate_detection/malaysia_plate_v5/weights/best.pt")
model.val(data="../plate-dataset/data.yaml")
```
