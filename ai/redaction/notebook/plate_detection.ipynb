```python
"""Face detection module using RetinaFace."""

from pathlib import Path

import torch
import wandb
from ultralytics import YOLO
```

```python
def main():
    """Train a YOLOv8 model for license plate detection using a custom dataset."""
    dataset_path = Path("../plate_dataset/data.yaml")
    "/Users/amber/Desktop/hackathon/disaster-readiness/ai/redaction"
    model = YOLO("yolov8m.pt")

    config = {
        "model":"yolov8m.pt",
        "epochs": 50,
        "img_size": 640,
        "batch_size": 16,
        "device": "cuda" if torch.cuda.is_available() else "cpu"
    }

    wandb.init(
        project="plate-detector-model",
        name="version1",
        config=config,
    )

    model.train(
        data=str(dataset_path),
        epochs=config["epochs"],
        imgsz=config["img_size"],
        batch=config["batch_size"],
        device=config["device"],
        exist_ok=True,
    )

    best_model_path = model.best.pt

    wandb.finish()
if __name__ == "__main__":
    main()
```
